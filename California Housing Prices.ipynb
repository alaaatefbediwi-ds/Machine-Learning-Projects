```python
import pandas as pd
```

```python
df = pd.read_csv(r"E:\Python\PythonDatasets\creditDetection\califroniaHouses\housing.csv")
df
```

```python
df.isnull().sum()
```

```python
# Calculating the mean of column 'total_bedrooms'
mean_bedrooms = df['total_bedrooms'].mean()
mean_bedrooms
```

```python
# Replacing the null values in column 'total_bedrooms' with their mean
df['total_bedrooms'].fillna(mean_bedrooms, inplace=True)
df
```

```python
#Separate data into training and testing
#Shuffling data frame so the order of the rows is randomized
df = df.sample(frac=1, random_state=2)
train_df = df[:17000]
#removing the existing index and replaces it with a new sequential index starting from 0
train_df = train_df.reset_index(drop=True)

test_df = df[17000:]
test_df = test_df.reset_index(drop=True)

train_df
```

```python
test_df
```

```python
y_train = train_df['median_house_value'].to_numpy()
y_test = test_df['median_house_value'].to_numpy()

y_train.shape, y_test.shape
```

# Baseline Model

```python
from sklearn.metrics import mean_absolute_error
```

```python
average_median_house_value = train_df['median_house_value'].mean()
baseline_model_test_predictions = [average_median_house_value] * len(test_df)

mean_absolute_error(baseline_model_test_predictions, y_test)
```

```python
# Calculating the error of random forest on (x_train_3, y_train) and testing on (x_test_3, y_test)
forest_base = RandomForestRegressor(n_estimators=50, max_depth=5).fit(x_train_3, y_train)
forest_base_test_predictions = forest_base.predict(x_test_3)
mean_absolute_error(y_test, forest_base_test_predictions)
```

# Dimensionality Reduction (PCA)

```python
from sklearn.decomposition import PCA
```

```python
x_train_pca = pca.transform(x_train_3)
x_test_pca = pca.transform(x_test_3)
x_train_pca.shape, x_test_pca.shape
```

```python
# Calculating the error of random forest on (x_train_pca, y_train) and testing on (x_test_pca, y_test)
forest_base = RandomForestRegressor(n_estimators=50, max_depth=5).fit(x_train_pca, y_train)
forest_base_test_predictions = forest_base.predict(x_test_pca)
mean_absolute_error(y_test, forest_base_test_predictions)
```

# Creating a Pipeline (Standardization => PCA=> RandomForest)

```python
from sklearn.pipeline import Pipeline
```

```python
scale_pca_pipe_forest = Pipeline(steps=[('Scaler', StandardScaler()),
                                       ('PCA', PCA(n_components=2)),
                                       ('Forest', RandomForestRegressor(n_estimators=50, max_depth=5))])
scale_pca_pipe_forest
```

```python
#Calculating error of Standardization => PCA => Random Forest pipeline
scale_pca_pipe_forest.fit(x_train_3, y_train)
scale_pca_pipe_forest_test_predictions = scale_pca_pipe_forest.predict(x_test_3)

mean_absolute_error(y_test, scale_pca_pipe_forest_test_predictions)
```

# K-means Clustering

```python
import matplotlib
```

```python
plt.scatter(train_df['longitude'], train_df['latitude'])
```

```python
from sklearn.cluster import KMeans
```

```python
x_train_lat_long = train_df[['longitude', 'latitude']].to_numpy()
kmeans = KMeans(n_clusters=3).fit(x_train_lat_long)
kmeans.labels_
```

```python
#visualizing Colored-map 
import plotly.express as px
```

```python
px.scatter(x=train_df['longitude'], y=train_df['latitude'], color=kmeans.labels_)
```

```python
# x_train clustering
x_train_clustering = pd.get_dummies(pd.Series(kmeans.labels_)).to_numpy()
x_train_clustering
```

```python
# Convert True and False to 1 and 0
x_train_clustering = x_train_clustering.astype(int)
x_train_clustering
```

```python
# predicting the clusters for the test data 
x_test_lat_long = test_df[['longitude', 'latitude']].to_numpy()
x_test_clustering = pd.get_dummies(pd.Series(kmeans.predict(x_test_lat_long))).to_numpy()
x_test_clustering.shape
```

```python
# checking the error
linear_clustering = LinearRegression().fit(x_train_clustering, y_train)
linear_clustering_test_predictions = linear_clustering.predict(x_test_clustering)

mean_absolute_error(linear_clustering_test_predictions, y_test)
```
